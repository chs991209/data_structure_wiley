위에서 정의한 ADT MyList를 바탕으로 Class LinkedList를 아래와 같이 정의하고자 한다.
이때 Class Node를 아래와 같이 정의하였을 때 아래의 methods getItem, setItem,
insertItem, removeItem, printMyList를 완성하고 LinkedList를 create 하고 items “4”, “3”과
“1”을 list의 시작하는 곳 (position 0)에 insert 하고, item “2”를 list의 position 1에 insert한
후, list의 내용을 출력하시오. 
또한 list의 position 3과 1의 item을 list에서 제거하고, 
그리고 첫번째 item (position 0)를 제거한 후 list의 내용을 출력하시오.

```python
class Node(object):
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next_node = next_node

    def get_data(self):
        return self.data

    def set_data(self, data):
        self.data = data

    def get_next(self):
        return self.next_node

    def set_next(self, new_next):
        self.next_node = new_next


class LinkedList(object):
    def __init__(self, head=None):
        self.head = head
        self.length = 0

    def __len__(self):
        return self.length

    def get_item(self, j):
        if j < 0 or j >= self.length:
            raise IndexError('Index out of the range of the sequence')
        current_node = self.head
        for k in range(j):
            current_node = current_node.get_next()
        return current_node.get_data()

    def set_item(self, val, j):
        if j < 0 or j >= self.length:
            raise IndexError('Index out of the range of the sequence')

        #  if j < -self.length or j >= self.length:
        #     raise IndexError("Sequence index out of range.")
        # elif -self.length <= j < 0:
        #     j += self.length  
        current_node = self.head
        # current_node를 head에서부터 한 칸씩 이동합니다. j-1에서 한 번 더 이동.
        for k in range(j):
            current_node = current_node.get_next()
        current_node.set_data(val)

    def insert_item(self, item, j=0):
        if j < 0 or j > self.length:
            raise IndexError('Index out of the range of the sequence')

        if j == 0:
            new_node = Node(item, self.head)
            self.head = new_node
        else:
            current_node = self.head
            for k in range(j - 1):
                current_node = current_node.get_next()
            new_node = Node(item, current_node.get_next())
            current_node.set_next(new_node)

        self.length += 1

    def remove_item(self, j=0):
        if j < 0 or j >= self.length:
            raise IndexError('Index out of range of the sequence')
        if j == 0:
            self.head = self.head.get_next()
        else:
            current_node = self.head
            for k in range(j - 1):
                current_node = current_node.get_next()
            current_node.set_next(current_node.get_next().get_next())
        self.length -= 1

    def print_list(self):
        current_node = self.head
        # current_node가 None이 될 때까지 LinkedList를 순회합니다.
        while current_node:
            print(current_node.get_data(), end=" ")
            current_node = current_node.get_next()
        print()

```

위에서 정의한 ADT MyList MyList를 바탕으로 
Class DoublyLinkedList DoublyLinkedList를 아래와 같이 정의하고자 한다 
이때 Class Node Node를 아래와 같이 정의하였을 때 Node Class와 DoublyLinkedList를 완성하세요 DoublyLinkedList를 create 하고 items “4”, “3” 과 "1" 을
list의 시작하는 곳 (position 0) 에 insert하고 
item “2” 를 list의 position 1에 insert한 후 list의 내용을 출력하시오
또한 list의 position 3과 1의 item을 list에서 제거하고 그리고 첫번째 item (position 0)
를 제거한 후 list의 내용을 출력하시오

```python
from abc import ABCMeta, abstractmethod


class MyList(object):
    """Our own version of collections.Sequence abstract base class."""
    __metaclass__ = ABCMeta

    @abstractmethod
    def __len__(self):
        pass

    @abstractmethod
    def __getitem__(self, j):
        pass

    @abstractmethod
    def __setitem__(self, val, j):
        pass

    @abstractmethod
    def insert_item(self, item, j=0):
        """Add an item into the List at the jth position."""
        pass

    @abstractmethod
    def remove_item(self, j=0):
        """Remove the item from the List at the jth position."""
        pass

    @abstractmethod
    def print_mylist(self):
        """Print the items in the list."""
        pass


class Node(object):
    def __init__(self, data=None, next_node=None, previous_node=None):
        self.data = data
        self.next_node = next_node
        self.previous_node = previous_node

    def get_data(self):
        return self.data

    def set_data(self, data):
        self.data = data

    def get_next(self):
        return self.next_node

    def set_next(self, new_next):
        self.next_node = new_next

    def get_previous(self):
        return self.previous_node

    def set_previous(self, new_previous):
        self.previous_node = new_previous


class DoublyLinkedList(MyList):
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def __len__(self):
        return self.length

    def insert_item(self, item, j=0):
        new_node = Node(data=item)
        # 기존에 아무것도 없을 때
        if self.head is None:
            self.head = new_node
            self.tail = new_node

            self.length += 1

        # head에 삽입하는 경우
        elif j == 0:
            # new_node.next_node는 self.head 
            new_node.set_next(self.head)
            self.head.set_previous(new_node)
            self.head = new_node
            self.length += 1
        else:
            current_node = self.head
            for k in range(j - 1):
                if current_node is None:
                    # head ~ index j - 1까지 전부 None이 아닐 시 통과, 적어도 j가 tail이어야 합니다.
                    raise IndexError('Index out of the range of the  doublylinkedlist')
                # current_node 가 j - 1이 되면 loop를 종료합니다.
                current_node = current_node.get_next()

            # j - 1 
            new_node.set_next(current_node.get_next())  # new_node의 next는 기존의 j node
            # new_node가 아닌 tail이 남을 경우
            # tail의 previous => new node,

            if current_node.get_next() is not None:
                # 기존의 j가 가리키던 node가 new_node를 previous로 참조하도록 합니다.
                current_node.get_next().set_previous(new_node)
            current_node.set_next(new_node)  # j가 new_node를 가리키도록 합니다.
            new_node.set_previous(current_node)  # new_node의 previous가 기존 j-1이 가리키던 node가 되도록 합니다.
            if new_node.get_next() is None:
                self.tail = new_node
        self.length += 1

    def remove_item(self, j=0):
        if self.head is None:
            raise IndexError('The list is empty')
        if j == 0:
            if self.head.get_next() is None:
                self.head = None
                self.tail = None
                self.length -= 1
            else:
                # head를 한 개 뒤로 옮깁니다.
                self.head = self.head.get_next()
                # head의 previous의 pointer가 None을 가리키도록 합니다.
                self.head.set_previous(None)
                self.length -= 1
        else:
            # head 포인터가 지정하는 node를 할당.
            current_node = self.head
            for k in range(j):
                if current_node is None:
                    raise IndexError('Index out of range')
                current_node = current_node.get_next()
            if current_node is None:
                raise IndexError('Index out of range')

            if current_node.get_previous() is not None:
                # j - 1 의 next node를 j + 1로 새로 할당합니다.
                current_node.get_previous().set_next(current_node.get_next())
            if current_node.get_next() is not None:
                # j + 1 의 previous가 j - 1이 되도록합니다.
                current_node.get_next().set_previous(current_node.get_previous())

            # current_node가 j != 0이더라도, None, el1, el2, .. 형태이며 j = 1이더라도 head를 가리킬 수 있다.
            # current_node.get_previous() is not None일 때는 current_node (j pointing)가 head가 아닌 겅우이므로
            # head == j node, head !== j node로 분기처리된 것과 같다.
            if current_node == self.head:
                self.head = current_node.get_next()
            if current_node == self.tail:
                # tail을 제거하면 이전의 node를 self.tail에 할당
                # singly linked list는 불가능
                self.tail = current_node.get_previous()
        self.length -= 1

    def print_list(self):
        current_node = self.head
        while current_node is not None:
            print(current_node.get_data(), end=" ")
            current_node = current_node.get_next()
        print()
```

```python
list_sample = DoublyLinkedList()
list_sample.insert_item("4", j=0)
list_sample.insert_item("3", j=0)
list_sample.insert_item("1", j=0)
list_sample.insert_item("2", j=1)
list_sample.print_list()

list_sample.remove_item(3)
list_sample.remove_item(1)
list_sample.remove_item(0)
list_sample.print_list()
```

```python
from abc import ABCMeta, abstractmethod


class MyQueue(object):
    """Our own version of collections.Sequence abstract base class."""
    __metaclass__ = ABCMeta

    @abstractmethod
    def len(self):
        """Returns the number of elements stored."""
        pass

    @abstractmethod
    def first(self):
        """Returns the element at the front without removing it."""
        pass

    @abstractmethod
    def is_empty(self):
        """Returns whether no elements are stored."""
        pass

    @abstractmethod
    def enqueue(self, item):
        """Add an item into the queue."""
        pass

    @abstractmethod
    def dequeue(self):
        """Remove an item from the queue."""
        pass

    @abstractmethod
    def print_myqueue(self):
        """Print the items in the queue."""
        pass


class Node(object):
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next_node = next_node

    def get_data(self):
        return self.data

    def get_next(self):
        return self.next_node

    def set_next(self, new_next):
        self.next_node = new_next


class LinkedListBasedQueue(MyQueue):
    def __init__(self, head=None, tail=None):
        self.head = None
        self.tail = None
        self.length = 0

    def len(self):
        return self.length

    def first(self):
        if self.head is None:
            raise ValueError('Queue is Empty')
        return self.head.get_data()

    def is_empty(self):
        # return self.tail is None
        return self.length == 0

    def enqueue(self, item):
        new_node = Node(data=item)
        if self.length == 0:
            self.head = new_node

        else:
            self.tail.set_next(new_node)

        self.tail = new_node
        self.length += 1

    def dequeue(self):
        # if self.head and self.tail is None:
        #     raise ValueError('Queue is Empty')
        if self.is_empty():
            raise ValueError('Queue is Empty')
        head_data = self.head.get_data()
        self.head = self.head.get_next()
        if self.head is None:
            self.tail = None
        self.length -= 1
        return head_data

    def print_myqueue(self):
        current_node = self.head
        # for _ in range(self.length):
        #     print(current_node.get_data(), end=' ')
        #     current_node = current_node.get_next()
        while current_node is not None:
            print(current_node.get_data(), end=' ')
            current_node = current_node.get_next()
        print()
```
