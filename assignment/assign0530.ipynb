```python
from bintrees import BinaryTree, AVLTree

# Tree Traversal functions declaration using recursion
# Additionally used closure of python to reduce using global objects(of python).
# ex) preorder()는 _preorder()를 내장함으로써 불필요한 global function 할당을 줄입니다.
# python closure using -> shows encapsualtion of python
def inorder(tree):
    return [key for key in tree.keys()]

def preorder(tree):
    result = []
    def _preorder(node):
        if node:
            result.append(node.key)
            _preorder(node.left)
            _preorder(node.right)
    _preorder(tree._root)
    return result

def postorder(tree):
    result = []
    def _postorder(node):
        if node:
            _postorder(node.left)
            _postorder(node.right)
            result.append(node.key)
    _postorder(tree._root)
    return result



# Quest1: Binary Tree Insertion for the sequence below
bst = BinaryTree()
insert_sequence = [5, 3, 7, 8, 2, 9, 1]
for num in insert_sequence:
    bst.insert(num, None)

print("1. BST Insertion")
print("In-order:", inorder(bst))
print("Pre-order:", preorder(bst))
print("Post-order:", postorder(bst))

# Quest 2: BST Deletion as below
delete_sequence = [1, 2, 5, 8, 9]
for num in delete_sequence:
    bst.remove(num)

print("\n2. BST Deletion")
print("In-order:", inorder(bst))
print("Pre-order:", preorder(bst))
print("Post-order:", postorder(bst))

# Quest 3: AVL Tree Insertion as below
avl = AVLTree()
insert_sequence = [5, 3, 7, 8, 2, 9, 1]
for num in insert_sequence:
    avl.insert(num, None)

print("\n3. AVL Tree Insertion")
print("In-order:", inorder(avl))
print("Pre-order:", preorder(avl))
print("Post-order:", postorder(avl))

# Quest 4: AVL Tree Deletion as below
delete_sequence = [1, 2, 5, 8, 9]
for num in delete_sequence:
    avl.remove(num)

print("\n4. AVL Tree Deletion")
print("In-order:", inorder(avl))
print("Pre-order:", preorder(avl))
print("Post-order:", postorder(avl))
```

```python
from bintrees import BinaryTree, AVLTree

# Additionally used closure of python to reduce using global objects(of python).
# ex) preorder()는 _preorder()를 내장함으로써 불필요한 global function 할당을 줄입니다.
# python closure using -> shows encapsualtion of python
def inorder(tree):
    """
    Perform in-order traversal on a binary search tree or AVL tree.
    
    Parameters:
    tree (BinaryTree or AVLTree): The tree to traverse.
    
    Returns:
    list: A list of keys in in-order traversal order.
    """
    return [key for key in tree.keys()]

def preorder(tree):
    """
    Perform pre-order traversal on a binary search tree or AVL tree.
    
    This function demonstrates the use of recursion and Python closures.
    By nesting the helper function `_preorder` within `preorder`, we reduce
    unnecessary global function allocations and encapsulate the helper function.
    
    Parameters:
    tree (BinaryTree or AVLTree): The tree to traverse.
    
    Returns:
    list: A list of keys in pre-order traversal order.
    """
    result = []
    def _preorder(node):
        """
        Helper function to recursively perform pre-order traversal.
        
        Parameters:
        node (TreeNode): The current node being visited.
        """
        if node:
            result.append(node.key)
            _preorder(node.left)
            _preorder(node.right)
    _preorder(tree._root)
    return result

def postorder(tree):
    """
    Perform post-order traversal on a binary search tree or AVL tree.
    
    This function also uses recursion and demonstrates Python closures.
    By nesting the helper function `_postorder` within `postorder`, we reduce
    unnecessary global function allocations and encapsulate the helper function.
    
    Parameters:
    tree (BinaryTree or AVLTree): The tree to traverse.
    
    Returns:
    list: A list of keys in post-order traversal order.
    """
    result = []
    def _postorder(node):
        """
        Helper function to recursively perform post-order traversal.
        
        Parameters:
        node (TreeNode): The current node being visited.
        """
        if node:
            _postorder(node.left)
            _postorder(node.right)
            result.append(node.key)
    _postorder(tree._root)
    return result

# Example usage to verify correctness (not part of the function docstrings)
if __name__ == "__main__":
    # Problem 1: BST Insertion
    bst = BinaryTree()
    insert_sequence = [5, 3, 7, 8, 2, 9, 1]
    for num in insert_sequence:
        bst.insert(num, None)

    print("1. BST Insertion")
    print("In-order:", inorder(bst))
    print("Pre-order:", preorder(bst))
    print("Post-order:", postorder(bst))

    # Problem 2: BST Deletion
    delete_sequence = [1, 2, 5, 8, 9]
    for num in delete_sequence:
        bst.remove(num)

    print("\n2. BST Deletion")
    print("In-order:", inorder(bst))
    print("Pre-order:", preorder(bst))
    print("Post-order:", postorder(bst))

    # Problem 3: AVL Tree Insertion
    avl = AVLTree()
    insert_sequence = [5, 3, 7, 8, 2, 9, 1]
    for num in insert_sequence:
        avl.insert(num, None)

    print("\n3. AVL Tree Insertion")
    print("In-order:", inorder(avl))
    print("Pre-order:", preorder(avl))
    print("Post-order:", postorder(avl))

    # Problem 4: AVL Tree Deletion
    delete_sequence = [1, 2, 5, 8, 9]
    for num in delete_sequence:
        avl.remove(num)

    print("\n4. AVL Tree Deletion")
    print("In-order:", inorder(avl))
    print("Pre-order:", preorder(avl))
    print("Post-order:", postorder(avl))
```
